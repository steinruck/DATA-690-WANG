```
# create new variable with desired column names
columns = ['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']

# assign all data from those columns to new dataframe
df1 = pd.DataFrame(df, columns=columns)

# print first line to verify
df1.head(1)
```

# Dataframe Key:

- TUITIONFEE_IN = In-state tuition and fees
- MN_EARN_WNE_P10 = Mean earnings of students working and not enrolled 10 years after entry

```
# find and display UMBC
df3 = df1[df['INSTNM'].str.contains("University of Maryland-Baltimore County")]
df3
```

```
# Find Univ. of Maryland College Park
df4 = df1[df['INSTNM'].str.contains("University of Maryland-College Park")]
df4
```

```
# Find and display Loyola College Maryland
df5 = df1[df['INSTNM'].str.contains("Loyola University Maryland")]
df5
```

```
# Find and display Johns Hopskins
df6 = df1[df['INSTNM'].str.contains("Johns Hopkins")]
df6
```

```
frames = [df3, df4, df5, df6]

df7 = pd.concat(frames)

df7
```

```
# assign in state tuition column to variable
# Calculate the min, max, and average in-state tution of the above four institutions.

tuition = df7["TUITIONFEE_IN"]

max_value = tuition.max()
min_value = tuition.min()
average_value = tuition.mean()
```

```
# print min, max, average

print("Minimum In-State Tuition =", min_value)
print("Maximum In-State Tuition =", max_value)
print("Average In-State Tuition =", average_value)
```
