```
import pandas as pd
import matplotlib.pyplot as plt
```

```
# ax = axis, fig = figure
# 2,2 makes 4 empty graphs
fig, ax_list = plt.subplots(2,2,figsize=(8,6))
fig.show()
```

```
x = [1,2,3,4]
y = [10,15.5,-12,34]

# 0,0 referencecs the first graph
# have to assign the data and the graphs in the same code
fig, ax_list = plt.subplots(2,2,figsize=(8,6))
ax_list[0,0].plot(x,y)
fig.show()
```

```
x = [1,2,3,4]
y = [10,15.5,-12,34]

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x,y)
fig.show()
```

```
x = [1,2,3,4]
y = [10,15.5,-12,34]

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x,y, 'ro') # r stands for red, 0 for dot. Also have x and s for square
fig.show()
```

```
x = [1,2,3,4]
y = [10,15.5,-12,34]

fig, ax = plt.subplots(figsize=(12,6)) # 12 maxes it bigger
ax.plot(x,y, 'ks')
ax.set_title("And Example of Matplotlib")
ax.set_xlabel("Height")
ax.set_ylabel("Weight")
fig.show()
```

```
x = [6.1,5.8,5.7,5.7,5.8,5.6,5.5,5.3,5.2,5.2]
y = [3.4,3.5,3.5,4.1,5.1,5.2,5.3,5.9,6.0,6.1]

fig, ax = plt.subplots(figsize=(12,6)) # 12 maxes it bigger
ax.plot(x,y, 'ks')
ax.set_title("Relationship Between Unemployment and Stock Price")
ax.set_xlabel("Unemployment")
ax.set_ylabel("Stock Price")
fig.show()
```

```
Country = ['USA', 'Canada', "Germany", 'UK', 'France']
GDP_per_capita = [45000, 42000, 52000, 49000, 47000]

fig, ax = plt.subplots(figsize=(12,6)) # 12 maxes it bigger

ax.bar(Country, GDP_per_capita)
ax.set_title("GDP Per Capita of Developed Countries")
ax.set_xlabel("Country")
ax.set_ylabel("per Capita GDP")
fig.show()
```

```
fig.savefig("gdp.png", dpi=300) # save figure. dpi changes resolution
```
